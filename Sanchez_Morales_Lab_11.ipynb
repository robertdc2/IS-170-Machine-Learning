# Lab 11 SVM

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn import metrics
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
```

```python
mydata=pd.read_csv("Heart_disease_statlog.csv")
```

```python
mydata.head()
```

```python
mydata['target']=pd.factorize(mydata['fbs'])[0]
```

```python
mydata.head()
```

```python
mydata2=mydata.drop('fbs',axis=1)
```

```python
mydata2.head()
```

```python
x=mydata2.drop('target',axis=1)
y=mydata['target']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=5)
```

```python
#kernel specifying. 
#There are linear (most common), Gaussian, polynomial, sigmoid, and computable kernel.
#Always try linear first and then try the others (again "trial & error" rule)
#'rbf' stands for radial basis function kernel which is the second best choice after linear

clf=svm.SVC(kernel='rbf')

#train and fit the model
clf.fit(x_train,y_train)
```

```python
#predict the response
y_pred=clf.predict(x_test)
```

```python
print("Accuracy: ", metrics.accuracy_score(y_test,y_pred))
```

```python
#the default average value of both precision and recall is 'binary' which is incorrect here as 
#we have three target labels. When there are multi labels in target, use 'macro' option. 
print("Precision: ",metrics.precision_score(y_test,y_pred, average='macro'))

print("Recall: ",metrics.recall_score(y_test,y_pred, average='macro'))
```
