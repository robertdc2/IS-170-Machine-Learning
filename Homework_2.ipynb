```python
import pandas as pd
import numpy as np

df=pd.read_csv("Food_Truck_Zones.csv")
df
```

```python
df['Shape.STArea()']=df['Shape.STArea()'].replace(0,df['Shape.STArea()'].mean())
df
```

```python
df['Shape.STLength()']=df['Shape.STLength()'].replace(0,df['Shape.STLength()'].mean())
df
```

```python
df['OBJECTID']=df['OBJECTID'].replace(0,df['OBJECTID'].mean())
df
```

```python
from sklearn.preprocessing import LabelEncoder
le= LabelEncoder()
label = le.fit_transform(df['Authorization'])
label
```

```python
df.drop('Authorization', axis=1, inplace=True)
df['Authorization'] = label
df
```

```python
from sklearn import preprocessing
df_scaled = preprocessing.scale(df)
df_scaled=pd.DataFrame(df_scaled,columns=df.columns)
df_scaled['the_geom'] = df['the_geom']
df=df_scaled
```

```python
old_cols=df.columns.values
new_cols=['the_geom', 'OBJECTID', 'Zoning Code', 'Authorization', 'Source', 'Food Truck Zone', 'Shape.STArea()', 'Shape.STLength()']
df=df.reindex(columns=new_cols)
df
```

```python
print(df.rename(columns={'Authorization' : 'Permission', 'Food Truck Zone': 'Permitted or Franchised'}))
```
