# **Association Rules**


Data Set #1 Zoo animals

```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
# created new header called zoo animals
df=pd.read_csv('zoo.csv', names=[' zoo animals '])
df.head() 
```

```python
df.shape
```

```python
#the following code converts a pandas dataframe into a list of lists.
data=list(df[" zoo animals "].apply(lambda x:x.split(",")))
data
```

```python
#Let's transform the list with one-hot encoding (make sure you understand the one-hot encoding concept first)
#Each column represents an item and each row represent a record or a transaction. 
#Initially, if that particular item occurs then it's labeled TRUE, if not then FALSE.
#After that, TRUE is converted to 1 and FALSE to 0 because computer loves numeric values.

from mlxtend.preprocessing import TransactionEncoder
a=TransactionEncoder()
a_data=a.fit(data).transform(data)
df=pd.DataFrame(a_data, columns=a.columns_)

#Turn False & True to 0 & 1 for easy readability
df=df.replace(False,0)
df=df.replace(True,1)

df
```

```python
#We'll try 0.1 for the SUPPORT threshold value 
# The parameter (use_colnames=True) is used to convert the integer back to its respective name.
df=apriori(df,min_support=0.1,use_colnames=True)
df
```

```python
#Let's view our interpretation values using the Association rule function.
df_ar=association_rules(df,metric="lift", min_threshold=1)
df_ar
```

```python
#Let's view our interpretation values using the Association rule function.
df_ar=association_rules(df,metric="confidence", min_threshold=0.6)
df_ar
```

From above result we could not find a correlation of the relationship between the bird, fish and mammal since they appear to share different characteristics.
for this data set only a few item sets were calculated for the support values and showed no relationship value for confidence, lift and conviction

```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
df=pd.read_csv('dog_breeds.csv', names=[' names '])
df.head() 
```

```python
df.shape
```

```python
data=list(df[" names "].apply(lambda x:x.split(",")))
data
```

```python
from mlxtend.preprocessing import TransactionEncoder
a=TransactionEncoder()
a_data=a.fit(data).transform(data)
df=pd.DataFrame(a_data, columns=a.columns_)

#Turn False & True to 0 & 1 for easy readability
df=df.replace(False,0)
df=df.replace(True,1)

df
```

```python
#attempted 0.2 for the SUPPORT threshold value 
df=apriori(df,min_support=0.2,use_colnames=True)
df
```

```python
#Let's view our interpretation values using the Association rule function.
df_ar=association_rules(df,metric="lift", min_threshold=1)
df_ar
```

```python
#Let's view our interpretation values using the Association rule function.
df_ar=association_rules(df,metric="confidence", min_threshold=0.6)
df_ar
```

From above result we could not find a correlation for the frequency of barking between different dog breeds. for this data set only three item sets were calculated for the support values and showed no relationship value for confidence, lift and conviction
