```python
import pandas as pd
import numpy as np

df=pd.read_csv("/content/Naive-Bayes-Classification-Data.csv")
df
```

```python
def merge (list1, list2):
  merged_list=tuple(zip(list1, list2))
  return merged_list

features=merge(df['glucose'], df['bloodpressure'])
features
```

```python
#Import Gaussian Naive Bayes model
from sklearn.naive_bayes import GaussianNB

#Create a Gaussian Classifier
model = GaussianNB()

# Train the model using the training sets
model.fit(features,df['diabetes'])

#Predict Output
predicted= model.predict([[0,2]]) # 0:No Diabetes, 1: Has diabetes
print ("Predicted Value:", predicted)
```

From the above result , 1 will indicate that the person does in fact have diabetes.

Determining gender based on inputs. 

```python
df=pd.read_csv("/content/Adult Census Income.csv")
df.head()
```

```python
#In order to process we must turn those categorical values into numberic values
#Import LabelEncoder
from sklearn import preprocessing

#creating labelEncoder
le = preprocessing.LabelEncoder()

#Using .fit_transform function to fit label
#encoder and return encoded label
sex_label=le.fit_transform(df['sex'])
df.drop("sex", axis=1, inplace=True)
df["sex"] = sex_label

marital_label=le.fit_transform(df['marital.status'])
df.drop("marital.status", axis=1, inplace=True)
df["marital.status"] = marital_label

relationship_label=le.fit_transform(df['relationship'])
df.drop("relationship", axis=1, inplace=True)
df["relationship"] = relationship_label


df
```

```python
df.columns= ['age', 'workclass', 'fnlwgt', 'education', 'education.num', 'occupation', 'race', 'capital.gain','capital.loss','hours.per.week', 'native.country', 'income', 'sex', 'marital.status','relationship']
df.head()
```

```python
#x_data = df[['age','education.num', 'fnlwgt', 'hours.per.week', 'marital.status','relationship']]
x_data = df[['age','education.num', 'hours.per.week', 'marital.status','relationship']]
y_data = df['sex']
#y_data = sex_label
```

```python
# Import train_test_split function
from sklearn.model_selection import train_test_split

# Split dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.3,random_state=109) # 70% training and 30% test
```

```python
#This is model generation

#Import Gaussian Naive Bayes model
from sklearn.naive_bayes import GaussianNB

#Create a Gaussian Classifier
model = GaussianNB()

#Train the model using the training sets
model.fit(X_train, y_train)

#Predict the response for test dataset
y_pred = model.predict(X_test)

#This is to evaluate the model
#Import scikit-learn metrics module for accuracy calculation
from sklearn import metrics

# Model Accuracy, how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```
